Definimos la funciones que utilizaremos para extraer los datos, ajustar la escala temporal y hacer las gráficas del campo magnético con respecto al tiempo, escalado y sin escalar

```python
#Función para graficar (coordenada temporal sin unidades)
def gsenzat(data,i,j):
    test = data[:,i,j]

    x = linspace(0, 397, 398)
    y = linspace(0, 198, 199)
    z = linspace(0, 206, 207)

    plot(z, test)
    xlabel("Coordenada Temporal sin Unidades")
    ylabel("Campo Magenitco [Gauss]")
```

```python
#Esta función convierte de segundos a minutos donde sea necesario; luego suma los intervalos para crear continuidad en el tiempo
def prep_time(times): 
    tmp = times.copy()    
    #Pasamos de segundos a minutos donde sea necesario.
    t = where(tmp==45.)
    for i in t:
        tmp[i] = tmp[i]/60.
        
    #Pasamos a horas.    
    for k in range(len(tmp)):
       tmp[k] = tmp[k]/60.
        
    #Sumamos los intervalos
    for j in range(len(times)-1):
       tmp[j+1] = tmp[j] + tmp[j+1]
    return tmp
```

```python
#Función para graficar con la escala temporal ajustada (horas)
def gcont(data, times, i, j):
    test = data[:,i,j]

    x = linspace(0, 397, 398)
    y = linspace(0, 198, 199)
    z = linspace(0, 206, 207)
 
    #Redimencionamos la coordenada temporal
    tmp = prep_time(times) #Preparamos los tiempos
        
    plot(tmp, test)
    xlabel("Tiempo[s]")
    ylabel("Campo Magenitco [Gauss]")
    show()
    scatter(tmp, test)
    xlabel("Tiempo[s]")
    ylabel("Campo Magenitco [Gauss]")
    #xlim(-100,)
    show()
    return tmp
```

```python
def likelihood(y_obs, y_model):
    chi_squared = (1.0/2.0)*sum(((y_obs-y_model)/y_model)**2)
    return exp(-chi_squared)
```

```python
def my_model(x_obs, m, b):
    return x_obs*m + b
```

```python
m_walk = empty((0)) #this is an empty list to keep all the steps
b_walk = empty((0))
l_walk = empty((0))

#randrange(start, stop[, step])

m_walk = append(m_walk, random.randn())
b_walk = append(b_walk, random.randn())

y_init = my_model(x_obs, m_walk[0], b_walk[0])
l_walk = append(l_walk, likelihood(y_obs, y_init))
print m_walk
print b_walk
print l_walk
```

```python
n_iterations = 20000 #this is the number of iterations I want to make
for i in range(n_iterations):
    m_prime = np.random.normal(m_walk[i], 0.1) 
    b_prime = np.random.normal(b_walk[i], 0.1)

    y_init = my_model(x_obs, m_walk[i], b_walk[i])
    y_prime = my_model(x_obs, m_prime, b_prime)
    
    l_prime = likelihood(y_obs, y_prime)
    l_init = likelihood(y_obs, y_init)
    
    alpha = l_prime/l_init
    if(alpha>=1.0):
        m_walk  = append(m_walk,m_prime)
        b_walk  = append(b_walk,b_prime)
        l_walk = append(l_walk, l_prime)
    else:
        beta = random.random()
        if(beta<=alpha):
            m_walk = append(m_walk,m_prime)
            b_walk = append(b_walk,b_prime)
            l_walk = append(l_walk, l_prime)
        else:
            m_walk = append(m_walk,m_walk[i])
            b_walk = append(b_walk,b_walk[i])
            l_walk = append(l_walk, l_init)
```
